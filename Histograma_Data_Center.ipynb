```python
# Carregar o dataset
df = pd.read_parquet("1_ano_dados_data_center.parquet")

# Definir a variável target
target_variable = 'requisicoes'

# Configurações para o plot
plt.figure(figsize=(12, 7))
sns.histplot(df[target_variable], kde=True, bins=50, color='skyblue')
plt.title(f'Distribuição de {target_variable.replace("_", " ").title()}', fontsize=16)
plt.xlabel(f'{target_variable.replace("_", " ").title()}', fontsize=12)
plt.ylabel('Frequência', fontsize=12)
plt.grid(axis='y', alpha=0.75)

# Salvar o histograma
plt.savefig('histograma_requisicoes.png')
plt.close()


print('Histograma salvo como histograma_requisicoes.png')
```
